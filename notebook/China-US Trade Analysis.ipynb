```python
# 分析准备：辅助画图的函数等

import pandas as pd

# 导入python画图库matplotlib
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm
% matplotlib inline

from matplotlib.pylab import mpl
# 指定默认字体
mpl.rcParams['font.sans-serif'] = ['FangSong'] 
mpl.rcParams['font.size']       = 12
# 解决保存图像是负号'-'显示为方块的问题
mpl.rcParams['axes.unicode_minus'] = False 

import numpy as np
from matplotlib.ticker import Formatter

class MyFormatter(Formatter):
    def __init__(self, dates, fmt='%Y%m'):
        self.dates = dates
        self.fmt = fmt

    def __call__(self, x, pos=0):
        """Return the label for time x at position pos"""
        ind = int(np.round(x))
        if ind >= len(self.dates) or ind < 0:
            return ''

        # return self.dates[ind].strftime(self.fmt)
        return pd.to_datetime(self.dates[ind], format="%Y-%m-%d").strftime(self.fmt)
```

```python
# 分析1 ： 中国 GDP 总量变化及三大产业的构成变化

import datetime

def draw_gdp_cn(df):
    
    # 设置图形大小
    fig = plt.figure(figsize=(14,18))

    ax = fig.add_subplot(3,1,1)
    
    # 画图 1
    axis_x = df.index
    idx = np.arange(len(axis_x))
    ax.plot(idx, df['gdp'].values)
    ax.plot(idx, df['gdp_ind_1'].values)
    ax.plot(idx, df['gdp_ind_2'].values)
    ax.plot(idx, df['gdp_ind_3'].values)
    ax.xaxis.set_major_formatter(MyFormatter(axis_x, '%Y-%m-%d'))
    ax.set_ylabel("GDP总量")
    legend = ['GDP总量', '第一产业', '第二产业', '第三产业']
    ax.legend(legend, loc="upper left")
    ax.set_title("中国名义GDP总量")
    
    # 画图 2
    ax2 = fig.add_subplot(3,1,2)
    ax2.plot(idx, df['gdp_avg'].values)
    ax2.xaxis.set_major_formatter(MyFormatter(axis_x, '%Y-%m-%d'))
    ax2.set_ylabel("人均GDP")
    ax2.set_title("人均GDP")

    # 画图 3
    ax3 = fig.add_subplot(3,1,3)
        
    tick_label = [x for x in axis_x]
    
    ax3.bar(idx, df['gdp_ind_1_pct'].values, tick_label = tick_label, label="第一产业")
    ax3.bar(idx, df['gdp_ind_2_pct'].values, bottom = df['gdp_ind_1_pct'].values, label="第二产业")
    ax3.bar(idx, df['gdp_ind_3_pct'].values, bottom = (df['gdp_ind_1_pct'] + df['gdp_ind_2_pct']).values, label="第三产业")
    ax3.set_ylabel("百分比")
    ax3.set_title('中国GDP构成')
    ax3.hlines(0.5, 0, 18, linestyle="--", linewidth=2, color="red")
    ax3.hlines(0.1, 0, 18, linestyle="--", linewidth=2, color="red")
    ax3.legend()
    
    plt.setp(plt.xticks()[1], rotation=45)
    
    #fig.autofmt_xdate()
    
    # 显示图片
    plt.show()   

draw_gdp_cn(df_cn_gdp_rmb)
```

```python
# 分析6：中国对外贸易情况 - 中美贸易组成情况（进出口）

def get_dist_data(df, date, dict_name):
    ts = df.loc[date]
    ts = ts.sort_values(ascending=False)
    values = ts.values

    total = ts.sum()

    table_vals = []
    for index, value in ts.iteritems():
        pct = "%.2f" % (float(value) / total * 100)
        name = dict_name[index]
        table_vals.append([index, name, pct])
    
    return values, table_vals

def draw_table(ax, col_labels, colWidths, table_vals):
    # 画表格
    table = ax.table(
                cellText=table_vals, 
                colLabels=col_labels,
                colWidths = colWidths,
                cellLoc='left',
                loc='bottom'
            )
    table.auto_set_font_size(False)
    table.set_fontsize(20)

    def set_pad_for_column(table, col, pad=0.1):
        cells = [key for key in table._cells if key[1] == col]
        for cell in cells:
            table._cells[cell].PAD = pad

    set_pad_for_column(table, col=1, pad=0.01)

    table.scale(1.5, 2)

def draw_pie(title, values, table_vals):
    # 设置图形大小
    fig, ax = plt.subplots(figsize=(16,10))

    # 画饼图 
    ax.pie(values, 
        #labels=labels, 
        autopct='%.1f%%', 
        startangle=90,
        radius = 20
    ) 

    # 设置样式
    ax.axis('equal')

    # 设置标题
    ax.set_title(title, loc="center")

    # 画表格
    col_labels = ['序列ID','出口项目','占比(%)']
    colWidths = [0.1, 0.7, 0.1]
    draw_table(ax, col_labels, colWidths, table_vals)
    
    # 调整子图之间的间距
    plt.tight_layout()

    # 显示图片
    plt.show()        
```

```python
# 单项进出口指标的变化情况

import numpy as np
import datetime
mpl.rcParams['font.size']       = 12

def convert_to_pct(df):
    df['sum'] = df.apply(np.sum, axis = 1)
    for col in df.columns:
        df[col] = df[col] / df['sum']
    
    return df

def draw_plot(seriesid, df, dict_name):
    # 设置图形大小
    fig, ax = plt.subplots(figsize=(16,8))

    # 画图 
    axis_x = [datetime.datetime.strptime(x, "%Y-%m-%d") for x in df.index]
    legend = [dict_name[seriesid]]
    ax.plot(axis_x, df[seriesid].values)

    ax.legend(legend, loc="upper left")
    
    # 显示图片
    plt.show()   
```
